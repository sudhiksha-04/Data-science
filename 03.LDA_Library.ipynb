```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
X1 = np.array( [[4,2],[2,4],[2,3],[3,6],[4,4],[1,2],[7,2],[3,1]] ) #class 1 datapoints 
X2 = np.array( [[9,10],[6,8],[9,5],[8,7],[10,8],[9,7],[6,9],[9,7]] ) #class 2 datapoints
```

```python
# Stack data together and create labels (0 for Class 1, 1 for Class 2)
X = np.vstack((X1, X2))  # Combine both class data into one data
y = np.array([0] * len(X1) + [1] * len(X2))  # Assign labels 
```

```python
lda = LinearDiscriminantAnalysis()  # Create an LDA object
lda.fit(X, y)  # Fit LDA to the dataset
```

```python
X_proj = lda.transform(X) 
```

```python
plt.scatter(X_proj[:len(X1)], np.zeros(len(X1)), color="red", marker="o", label="Class 1")  # Class 1 points
plt.scatter(X_proj[len(X1):], np.zeros(len(X2)), color="blue", marker="s", label="Class 2")  # Class 2 points
plt.axvline(x=0, color="black", linestyle="dashed", label="Decision Boundary")  # Vertical line as decision boundary

plt.xlabel("Projected Value")
plt.title("LDA Projection of Data")
plt.legend()
plt.show()
```

```python
new_sample = np.array([[5, 6]])  # Example point to classify
predicted_class = lda.predict(new_sample)  # Predict using LDA
print(f"Predicted class for {new_sample[0]} is {'Class 2' if predicted_class[0] == 1 else 'Class 1'}")
```
